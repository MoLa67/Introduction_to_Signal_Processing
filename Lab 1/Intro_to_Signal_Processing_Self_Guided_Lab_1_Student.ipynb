## Pre-lab theory

### 1) What is a signal?
A signal is a function that carries information. In signal processing we usually study:
- **Continuous-time signals**: defined for every time instant
- **Discrete-time signals**: defined only at sample points

### 2) Why Python?
Python is widely used for signal processing because it provides:
- `numpy` for numerical computation
- `matplotlib` for plotting
- `scipy.signal` for waveform generation and analysis

### 3) Common ideas used in this lab
- plotting continuous and discrete signals
- sampling and time indexing
- periodic and aperiodic behaviour
- even and odd symmetry
- basic signal operations such as shifting, scaling, addition, multiplication, and folding
