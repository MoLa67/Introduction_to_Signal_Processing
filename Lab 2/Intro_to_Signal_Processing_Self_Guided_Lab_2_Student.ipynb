## General Theory

Convolution is the mathematical operation used to determine the output of an LTI system from its input and impulse response. For a discrete-time system, the convolution sum combines one signal with a shifted version of another signal. For a continuous-time system, the convolution integral is approximated in this notebook using a sufficiently small sampling interval.

For an LTI system, the impulse response completely characterises the system. The step response describes how the system reacts to a unit step input. The frequency response shows how different frequency components are scaled and phase-shifted by the system.

# Practical 01 – Convolution of Discrete-Time Signals

### Theory

Discrete-time convolution is written as

$$
y[n] = x[n] * h[n] = \sum_{k=-\infty}^{\infty} x[k]h[n-k].
$$

It is widely used to analyse the output of discrete-time LTI systems.

# Practical 02 – Convolution of Continuous-Time Signals

### Theory

Continuous-time convolution is defined as

$$
y(t) = x(t) * h(t) = \int_{-\infty}^{\infty} x(\tau)h(t-\tau)\,d\tau.
$$

In numerical work, the integral can be approximated by sampling the signals with a small time step.

The frequency response of a discrete-time system is obtained by evaluating the transfer function on the unit circle:

$$
H(e^{j\omega}).
$$

Its magnitude shows how much each frequency is amplified or attenuated, while its phase shows the phase shift introduced by the system.
