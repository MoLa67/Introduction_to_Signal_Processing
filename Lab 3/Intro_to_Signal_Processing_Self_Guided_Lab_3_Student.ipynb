## Preliminary Theory

### 1. Signal Generation and Sampling
A signal is represented mathematically as a function of time. In computation, continuous signals are approximated by a dense time grid, while discrete-time signals are represented by sample values.

### 2. Superposition and LTI Systems
Linear time-invariant (LTI) systems obey superposition. Their response to an input can be studied using impulse response, convolution, and frequency-domain methods.

### 3. Fourier Transform Basics
The Fourier Transform describes how a signal is composed of sinusoidal frequency components. In this notebook, the engineering convention is used:

$$
X(\omega)=\int_{-\infty}^{\infty} x(t)e^{-j\omega t}dt
$$
